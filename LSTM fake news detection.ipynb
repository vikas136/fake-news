```python
x=df.iloc[:,:-1]
```

```python
y=df.iloc[:,-1]
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
voc_size=5000
```

```python
message=x.copy()
```

```python
message.reset_index(inplace=True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('all')
```

```python
sw=stopwords.words('english')
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
corpus=[]
```

```python
for i in range(len(message)):
    review=re.sub('[^a-zA-Z]',' ',message['Headline'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if word not in sw]
    review=' '.join(review)
    corpus.append(review)
```

```python
onehot=[one_hot(words,voc_size) for words in corpus]
onehot
```

```python
sent_length=20
```

```python
embedded_docs=pad_sequences(onehot,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
from tensorflow.keras.layers import Dropout
```

```python
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=[['accuracy']])
print(model.summary())
```

```python
import numpy as np
```

```python
x_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
x_final.shape
```

```python
y_final.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_final,y_final,test_size=0.33,random_state=43)
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),batch_size=64,epochs=10)
```

```python
from sklearn.metrics import *
```

```python
pred=model.predict_classes(x_test)
```

```python
cf=confusion_matrix(y_test,pred)
cf
```

```python
accuracy=accuracy_score(y_test,pred)
accuracy*100
```
