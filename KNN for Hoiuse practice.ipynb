```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:\\Users\\krishna\\AppData\\Local\\Temp\\Temp1_seattle-house-sales-prices.zip\\house_sales.csv')
df.head()
```

```python
df1=df._get_numeric_data()
df1
```

```python
df2=pd.DataFrame(data=df1,columns=('bedrooms','bathrooms','sqft_living','sqft_lot','sqft_above','price'))
df2
```

```python
x=df2.iloc[:,:-1].values
```

```python
y=df2.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train1=scaler.fit_transform(x_train)
x_test1=scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
for k in [1,2,3]:
    knr=KNeighborsRegressor(k)
    knn=knr.fit(x_train1,y_train)
    pred=knn.predict(x_test1)
```

```python
print("predicted value for k={}: {}".format(k,pred))
```

```python
from sklearn.metrics import mean_squared_error
from numpy import sqrt
```

```python
rmse_val=[]
for k in range(10):
    knr=KNeighborsRegressor(k)
    knr.fit(x_train1,y_train)
    pred1=knr.predict(x_test1)
    error=sqrt(mean_squared_error(y_test,pred1))
    rmse_val1.append(error)
    print('rmse value for k=',k,'is',error)
```
