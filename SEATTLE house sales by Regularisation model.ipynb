```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:\\Users\\krishna\\AppData\\Local\\Temp\\Temp1_seattle-house-sales-prices.zip\\house_sales.csv')
df
```

```python
df1=pd.DataFrame(data=df,columns=('sqft_living','bedrooms','sqft_lot','floors','grade','sqft_above','sqft_basement','lat','sqft_living15','sqft_lot15','price'))
df1                                  
```

```python
x=df1.iloc[:,:-1].values
```

```python
y=df1.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import Lasso
lasso=Lasso(alpha=0.2,normalize=False)
lasso.fit(x_train,y_train)
pred=lasso.predict(x_test)
pred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
from numpy import sqrt
rmse=sqrt(mean_squared_error(y_test,pred))
rmse
```

```python
from sklearn.metrics import r2_score
r2score=r2_score(y_test,pred)
accuracy=r2score*100
accuracy
```

```python
lasso.coef_
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge(alpha=0.5,normalize=False)
ridge.fit(x_train,y_train)
ridgepred=ridge.predict(x_test)
ridgepred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
from numpy import sqrt
rmse=sqrt(mean_squared_error(y_test,ridgepred))
rmse
```

```python
from sklearn.metrics import r2_score
r2score=r2_score(y_test,ridgepred)
accuracy=r2score*100
accuracy
```

```python
from sklearn.linear_model import ElasticNet
EN=ElasticNet(random_state=0)
EN.fit(x_train,y_train)
ENpred=EN.predict(x_test)
ENpred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
from numpy import sqrt
rmse=sqrt(mean_squared_error(y_test,ENpred))
rmse
```

```python
from sklearn.metrics import r2_score
r2score=r2_score(y_test,ENpred)
accuracy=r2score*100
accuracy
```
