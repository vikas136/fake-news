```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:\\Users\\krishna\\AppData\\Local\\Temp\\Temp1_creditcardfraud.zip\\creditcard.csv')
df
```

```python
x=df.iloc[:,:-1].values
```

```python
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import *
```

```python
from numpy import sqrt
```

```python
rmse=[]

for k in range(10):
    k=k+1
    knn=KNeighborsClassifier(k)
    knn.fit(x_train,y_train)
    pred=knn.predict(x_test)
    error=sqrt(mean_squared_error(y_test,pred))
    rmse.append(error)
    print('rmse value for k=',k,'is',error)
```

```python
curve=pd.DataFrame(rmse)
plt.plot(curve)
plt.show()
```

```python
knn1=KNeighborsClassifier(2)
knn1.fit(x_train,y_train)
pred1=knn1.predict(x_train)
pred1
```

```python
confusion=confusion_matrix(y_test,pred1)
confusion
```
