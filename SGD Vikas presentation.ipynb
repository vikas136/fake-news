```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('C:\\Users\\krishna\\AppData\\Local\\Temp\\Temp1_seattle-house-sales-prices.zip\\house_sales.csv')
df
```

```python
df1=pd.DataFrame(data=df,columns=('sqft_living','bedrooms','sqft_lot','floors','grade','sqft_above','sqft_basement','lat','sqft_living15','sqft_lot15','price'))
df1         
```

```python
x=df1.iloc[:,:-1].values
```

```python
y=df1.iloc[:,-1].values
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x1=ss.fit_transform(x)
x1
```

```python
df2=pd.DataFrame(data=x1,columns=('sqft_living','bedrooms','sqft_lot','floors','grade','sqft_above','sqft_basement','lat','sqft_living15','sqft_lot15'))
df2
```

```python
from sklearn.model_selection import train_test_split
x1_train,x1_test,y_train,y_test=train_test_split(x1,y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import SGDRegressor
sgd=SGDRegressor()
sgd.fit(x1_train,y_train)
pred=sgd.predict(x1_test)
pred
```

```python
new=sgd.predict([[0.12,0.96,0.87,0.67,0.98,0.84,0.56,0.87,0.64,0.76]])
new
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
from numpy import sqrt
rmse=sqrt(mean_squared_error(pred,y_test))
rmse
```

```python
from sklearn.metrics import r2_score
r2score=r2_score(pred,y_test)
Accuracy=r2score*100
Accuracy
```
