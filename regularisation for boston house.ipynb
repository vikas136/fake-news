```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import *
```

```python
boston=load_boston()
```

```python
x=boston['data']
```

```python
y=boston['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
df1=pd.DataFrame(boston['data'])
df1.columns=boston['feature_names']
df1['price']=boston['target']
```

```python
df1
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x1_train=ss.fit_transform(x_train)
x1_test=ss.transform(x_test)
```

```python
from sklearn.linear_model import Ridge
```

```python

ridge1=Ridge(alpha=0.5,normalize=True)
ridge1.fit(x1_train,y_train)
pred=ridge1.predict(x1_test)
pred
```

```python
ridge1.intercept_
```

```python
ridge1.coef_
```

```python
from sklearn.metrics import *
```

```python
accuracy=r2_score(y_test,pred)
accuracy*100
```

```python
from sklearn.linear_model import Lasso
lasso=Lasso(alpha=0.1,normalize=True)
lasso.fit(x1_train,y_train)
pred1=lasso.predict(x1_test)
pred1
```

```python
lasso.intercept_
```

```python
lasso.coef_
```

```python
accuracy=r2_score(y_test,pred1)
accuracy*100
```

```python
from sklearn.linear_model import ElasticNet
```

```python
elasticnet=ElasticNet(random_state=0)
elasticnet.fit(x1_train,y_train)
pred2=elasticnet.predict(x1_test)
```
