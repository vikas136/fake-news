```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:\\Users\\krishna\\Downloads\\titanic (1).csv')
df
```

```python
df1=pd.DataFrame(data=df,columns=['Age','Siblings/Spouses Aboard','Parents/Children Aboard','Fare','Pclass','Survived'])
df1
```

```python
x=df1.iloc[:,:-1].values
```

```python
y=df1.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x1_train=ss.fit_transform(x_train)
x1_test=ss.transform(x_test)
```

```python
from sklearn.svm import SVC
svc=SVC(kernel='rbf')
svc.fit(x1_train,y_train)
pred=svc.predict(x1_test)
pred
```

```python
new=svc.predict([[27,1,2,40,2]])
new
```

```python
from sklearn.metrics import *
```

```python
confusion=confusion_matrix(y_test,pred)
confusion
```

```python
print(classification_report(y_test,pred))
```
