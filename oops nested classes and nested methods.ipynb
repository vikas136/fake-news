```python
class results:
    def __init__(self,name,per,peri,perg):
        self.name=name
        self.ssc=self.ssc(per)
        self.inter=self.inter(peri)
        self.graduation=self.graduation(perg) 
    def displayresult(self):
        print('my name is:',self.name)
        self.ssc.sscv()
        self.inter.interv()
        self.graduation.gradper()
        
    
    class ssc:
        def __init__(self,per):
            self.per=per
            
        def sscv(self):
            print('my 10th percentage is:',self.per)
            
    class inter:
        def __init__(self,peri):
            self.peri=peri
            
        def interv(self):
            print('my inter percentage is:',self.peri)
            
    class graduation:
        def __init__(self,perg):
            self.perg=perg
        def gradper(self):
            print('my graduation percentage is:',self.perg)
n=int(input('enter number of students'))
for i in range(n):
    name=input('enter the name')
    per=int(input('enter the 10th marks'))
    peri=int(input('enter the inter marks'))
    perg=int(input('enter the graduation marks'))
    s=results(name,per,peri,perg)
    s.displayresult()
```
