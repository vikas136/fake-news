```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('C:\\Users\\krishna\\AppData\\Local\\Temp\\Temp1_graduate-admissions.zip\\Admission_Predict.csv')
df
```

```python
df1=df.set_index('Serial No.')
df1
```

```python
x=df1.iloc[:,:-1].values
```

```python
y=df1.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
DTR=DecisionTreeRegressor()
DTR.fit(x_train,y_train)
pred=DTR.predict(x_test)
pred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
from numpy import sqrt
rmse=sqrt(mean_squared_error(pred,y_test))
rmse
```

```python
from sklearn.metrics import r2_score
r2score=r2_score(pred,y_test)
accuracy=r2score*100
accuracy
```
